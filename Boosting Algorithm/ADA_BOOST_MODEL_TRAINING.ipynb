```python
# Importing necssary libraries
import pandas as pd
```

```python
# Loading the dataset
dataset = pd.read_csv("insurance_pre.csv")
dataset.head()
```

```python
# One-hot encoding to handle categorical data(state)
dataset = pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset.tail()    # displays only last 5 rows in the dataset
```

```python
dataset.columns
```

```python
# Feature selection or inputs
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent.head()
```

```python
# Target or output 
dependent = dataset[['charges']]
dependent.head()
```

```python
# Splitting data into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.25,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
# Model Training 
from sklearn.ensemble import AdaBoostRegressor
regressor = AdaBoostRegressor(random_state=0, n_estimators=100,learning_rate=1,loss='linear')
regressor.fit(X_train,Y_train)
```

```python
Y_pred = regressor.predict(X_test)
Y_pred
```

```python
# Model evaluation
from sklearn.metrics import r2_score
score = r2_score(Y_test,Y_pred)
score
```

```python
# Saving the model
import pickle
filename = "Random_Forest_Saved_Model.sav"
pickle.dump(regressor,open(filename,'wb'))
```
